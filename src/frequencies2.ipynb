```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = os.path.join('/home', 'jovyan', 'work', 'src', 'data', 'sample.xls')

if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path}")

data = pd.read_excel(file_path)

print(f"extracted data {data.head()}")
```

```python
prices = data['Price']

prices_mean = prices.mean()
prices_std = prices.std()
print(f"Mean price: {prices_mean}")
print(f"Standard deviation of prices: {prices_std}")
```

```python
prices_max = prices.max()
prices_min = prices.min()

print(f"Max price: {prices_max}")
print(f"Min price: {prices_min}")
```

```python
fig, ax = plt.subplots(figsize=(10,6))

counts, bins, patches = ax.hist(prices, color='lightblue', edgecolor='black')

bins_labels = [f"{int(bins[i])} - {int(bins[i+1])}" for i in range(len(bins) - 1)]


for i, (count, patch) in enumerate(zip(counts,patches)):
    height = patch.get_height()

    ax.text(
        patch.get_x() + patch.get_width() / 2,
        height + 0.5,
        f"{int(count)}",
        ha='center',
        va='bottom',
        fontweight='bold',
    )


ax.set_title('Price Distribution')

ax.set_xlabel('Price')
ax.set_ylabel('Frequency')

ax.set_xticks([(int(bins[i]) + int(bins[i+1])) / 2 for i in range(len(bins) - 1)])
ax.set_xticklabels(bins_labels, rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
