```python
import pandas as pd
import matplotlib.pyplot as plt
```


SkiLodges.com realiza una prueba de mercado de su nuevo sitio web y le interesa saber cuánta facilidad de navegación proporciona su diseño. Selecciona al azar 200 usuarios frecuentes de internet y les pide que lleven a cabo una búsqueda en la página web. A cada uno de ellos se le solicita calificar la relativa facilidad para navegar (mala, buena, excelente o sobresaliente). Los resultados aparecen en la siguiente tabla:

| **Calificación**   | **Frecuencia** |
|---------------------|----------------|
| Sobresaliente       | 102            |
| Excelente           | 58             |
| Buena               | 30             |
| Mala                | 10             |

---

### **Preguntas:**

1. Elabore una gráfica de barras con los resultados de la encuesta.  
1. Dibuje una gráfica de pastel con los resultados de la encuesta.  


```python
frecuency = pd.Series([102, 58, 30, 10])
calification = pd.Series(['Sobresaliente', 'Excelente', 'Buena', 'Mala'])
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.bar(calification, frecuency, color=['green', 'blue', 'orange', 'red'])

plt.subplot(1, 2, 2)
plt.pie(frecuency, labels=calification, autopct='%1.1f%%')

plt.suptitle('Valoración de mercado')
```
